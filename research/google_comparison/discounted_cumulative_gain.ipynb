```python
from collections import OrderedDict
import json
from pprint import pprint
from rank_metrics import dcg_at_k
```

```python
with open("data/queries-en.json") as en_queries_file:
    en_queries = json.load(en_queries_file)
with open("data/queries-nl.json") as nl_queries_file:
    nl_queries = json.load(nl_queries_file)
```

```python
for nl_query in nl_queries:
    print(nl_query["queries"][0])
    print("*" * len(nl_query["queries"][0]))
    sorted_items = sorted(
        ((item["hash"], item["rating"],) for item in nl_query["items"]), 
        key=lambda item: item[1],
        reverse=True  # TODO: check if higher is indeed better, it's required for the metric it seems
    )
    ranking = OrderedDict(sorted_items)
    pprint(ranking)
    print()
    print()
```
