```python
import os
import json
from functools import reduce
from bs4 import BeautifulSoup
from bs4.element import NavigableString
```

```python
def raw_html_document(file_path):
    with open(file_path) as fp:
        soup = BeautifulSoup(fp)
    texts = []
    for el in soup.descendants:
        if not isinstance(el, NavigableString):
            continue
        text = el.string.strip()
        texts.append(text)
    return texts
```

```python
for file in [file for file in os.scandir("data") if file.path.endswith("html")]:
    texts = raw_html_document(file.path)
    print("*" * 100)
    print()
    print()
    print("\n\n".join(texts))
    print()
    print()
```

```python
def update_periphery_texts(aggregation, file_path):
    with open(file_path) as fp:
        soup = BeautifulSoup(fp)
    texts = set()
    for el in soup.descendants:
        if not isinstance(el, NavigableString):
            continue
        text = el.string.strip()
        if text:
            texts.add(text)
    return aggregation & texts if aggregation else texts
```

```python
periphery_texts = set()
periphery_texts = reduce(
    update_periphery_texts, 
    [file.path for file in os.scandir("data") if file.path.endswith("html")], 
    periphery_texts
)
```

```python
def clean_html_document(file_path, periphery_texts, raw_texts):
    with open(file_path) as fp:
        soup = BeautifulSoup(fp)
    texts = []
    for el in soup.descendants:
        if not isinstance(el, NavigableString):
            continue
        text = el.string.strip()
        if text and text not in periphery_texts and isinstance(text, str) and text in raw_texts:
            texts.append(text)
    return texts
```

```python
for file in [file for file in os.scandir("data") if file.path.endswith("html")]:
    path, name = os.path.split(file.path)
    name, ext = os.path.splitext(name)
    with open(os.path.join(path, name + ".json")) as fp:
        tika = json.load(fp)
    texts = clean_html_document(file.path, periphery_texts, tika.get("text", ""))
    print("*" * 100)
    print()
    print()
    print("\n\n".join(texts))
    print()
    print()
```
